# Perceptron From Scratch

## Objective

The objective of this notebook is to understand the working of a Perceptron, which is the basic building block of Neural Networks.

A Perceptron takes multiple inputs, applies weights and bias, computes a weighted sum, passes it through an activation function, and generates an output.

This implementation is performed using pure Python without any Machine Learning or Deep Learning libraries.

## Problem Statement

A student has obtained marks in three subjects:

- Deep Learning = 75
- AWS = 65
- Data Engineering = 80

The perceptron will determine whether the student passes or fails based on the weighted sum of these marks.

Target Output:

- Pass = 1
- Fail = 0

```python
# Step 1: Input Data

x = [75, 65, 80]

print("Input Features:", x)
```

```python
# Step 2: Initialize Weights

w = [4, 3, 5]

print("Weights:", w)
```

```python
# Step 3: Initialize Bias

b = -800

print("Bias:", b)
```

```python
# Step 4: Learning Rate and Target Output

eta = 0.01

y = 1

print("Learning Rate:", eta)
print("Target Output:", y)
```

## Net Input Calculation

The net input is calculated using:

z = Σ(wᵢxᵢ) + b

where:

- x = input values
- w = weights
- b = bias

```python
# Step 5: Calculate Net Input

z = 0

for i in range(len(x)):
    z += x[i] * w[i]

z += b

print("Net Input (z):", z)
```

## Activation Function

A Step Activation Function is used.

If z >= 0:

Output = 1

Otherwise:

Output = 0

```python
# Step 6: Apply Activation Function

if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Predicted Output:", y_pred)
```

## Error Calculation

Error = Actual Output - Predicted Output

```python
# Step 7: Calculate Error

error = y - y_pred

print("Error:", error)
```

## Weight Update Rule

Perceptron Learning Rule:

Wnew = Wold + η × Error × X

If the error is zero, no update is required.

```python
# Step 8: Update Weights

for i in range(len(w)):
    w[i] = w[i] + eta * error * x[i]

print("Updated Weights:", w)
```

```python
# Step 9: Update Bias

b = b + eta * error

print("Updated Bias:", b)
```

```python
# Final Results

print("=" * 40)

print("Target Output    :", y)

print("Predicted Output :", y_pred)

print("Error            :", error)

print("Final Weights    :", w)

print("Final Bias       :", b)

print("=" * 40)
```

# Conclusion

A Perceptron was successfully implemented from scratch using pure Python.

The model calculated the weighted sum of inputs, applied a step activation function, generated a prediction, computed the error, and applied the perceptron learning rule.

Since the predicted output matched the target output, the error was zero and no weight update was required.

This experiment demonstrates the fundamental learning process used in neural networks and forms the foundation of Deep Learning.
